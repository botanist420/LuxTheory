```python
xy_col_list = ["toal_bill", "tip"]

# create a function: test_copilot_vis which turn the csv file into a dataframe: df_test
# and then print first 5 rows of this dataframe
# use seaborn to plot a scatter plot of the input columns names in xy_col_list
# use plt.show() to show the plot

def test_copilot_vis(csv_file, xy_col_list: list):
    # If the suffix of csv_file is not ".csv", then return "not a csv file" and exit the function
    if csv_file[-4:] != ".csv":
        return "not a csv file"
    else:
        df = pd.read_csv(csv_file)
        print(df.head())
        sns.scatterplot(data=df, x=xy_col_list[0], y=xy_col_list[1])
        plt.show()
    
```

```python
test_copilot_vis("tips.csv", ["total_bill", "tip"])
```
